product_data = {

````python
# Importer la bibliothèque pandas avec l'alias pd
import pandas as pd
````

````python
# données sous la forme de dictionnaire

product_data = {
    'product a': [13, 20, 0, 10], 
    'project b': [10, 30, 17, 20],
    'project c': [6, 9, 10, 0]
}
````

- Dans cette partie nous allons:

    * Appliquer quelques transformations au dataframe
    * Calculer quelques informations statistiques
 
 
 Essayer de décrire l'utilité de chaque opération et transformation :



````python
movies_df = pd.read_csv(r"C:\Users\ASUS\Downloads\IMDB-Movie-Data.txt", index_col="Title")
````

On affiche les 5 premiers lignes de movies_df : 

````python
movies_df.head()
````

On affiche les 2 derniers lignes de movies_df :

````python
movies_df.tail(2)
````

On affiche des informations sur movies_df :

````python
movies_df.info()
````

On affiche les dimensions de movies_df :

````python
movies_df.shape
````

On ajoute les lignes de temp_df à la fin de movies_df :

````python
temp_df = movies_df.append(movies_df)

temp_df.shape
````

On supprime les lignes dupliqués dans temp_df :

````python
temp_df = temp_df.drop_duplicates()

temp_df.shape
````

````python
temp_df.drop_duplicates(inplace=True)
````

On fait une nouvelle copie :

````python
temp_df = movies_df.append(movies_df)  # make a new copy

temp_df.drop_duplicates(inplace=True, keep=False)

temp_df.shape
````

On affiche les colonnes de movies_df : 

````python
movies_df.columns
````

On renomme les colonnes de movies_df :

````python
movies_df.rename(columns={
        'Runtime (Minutes)': 'Runtime', 
        'Revenue (Millions)': 'Revenue_millions'
    }, inplace=True)


movies_df.columns
````

On affiche les colonnes de movies_df :

````python
movies_df.columns = ['Rank', 'genre', 'description', 'director', 'actors', 'year', 'runtime', 
                     'rating', 'votes', 'revenue_millions', 'metascore']


movies_df.columns
````

````python
movies_df.columns = [col.lower() for col in movies_df]

movies_df.columns
````

On utilise la méthode isnull() pour retourner une valeur booléenne (False) qui indique si une expression ne contient pas de données valides (Null):

````python
movies_df.isnull()
````

On calcule la somme des valeurs nulles :

````python
movies_df.isnull().sum()
````

On supprime les valeurs manquantes :

````python
movies_df.dropna()
````

On supprime les lignes où tous les éléments sont nan :

````python
movies_df.dropna(axis=1)
````

On affiche la colonne 'revenue_millions' de movies_df :

````python
revenue = movies_df['revenue_millions']
````

On affiche les 5 premiers éléments de la colonne revenue :

````python
revenue.head()
````

On calcule la moyenne de revenue :

````python
revenue_mean = revenue.mean()

revenue_mean
````

On remplace les valeurs NULL par une valeur spécifiée :

````python
revenue.fillna(revenue_mean, inplace=True)
````

On calcule la somme des valeurs nulles :

````python
movies_df.isnull().sum()
````

On utilise la méthode describe() pour calculer certaines données statistiques comme le centile, la moyenne et la norme des valeurs numériques de la série ou du DataFrame :

````python
movies_df.describe()
````

On affiche movies_df triée selon la colonne rank :

````python
movies_df.sort_values(by='rank')
````
